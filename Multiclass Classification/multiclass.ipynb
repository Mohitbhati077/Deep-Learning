### Multiclass Classification


```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Creating a multiclass data

```python
from sklearn.datasets import make_classification
X,Y=make_classification(n_samples=6000,n_features=30,n_informative=15,n_classes=3,random_state=42)
```

```python
X
```

```python
Y
```

```python
X.shape
```

```python
Y.shape
```

### Convert x and y to dataframe

```python
cols=[f'x{i+1} 'for i in range(30)]
```

```python
cols
```

```python
import pandas as pd
X=pd.DataFrame(X,columns=cols)
X
```

```python
y=pd.DataFrame(Y,columns=['target'])
y
```

```python
y.value_counts()
```

### Create preprocessing pipeline for X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='median')),
                         ('scaler',StandardScaler())])
```

```python
X_pre=num_pipe.fit_transform(X)
X_pre
```

```python
cols=num_pipe.get_feature_names_out()
cols
```

```python
X_pre=pd.DataFrame(X_pre,columns=cols)
X_pre.head()
```

### Train test split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X_pre,y,test_size=0.2,random_state=42)
```

```python
xtrain.head()
```

```python
xtest.head()
```

### ONE HOT ENCODE Y feature

```python
from keras.utils import to_categorical
ytrain_cat=to_categorical(ytrain)
ytest_cat=to_categorical(ytest)
```

```python
ytrain_cat[0:5]
```

```python
ytest.head()
```

```python
ytrain.head()
```

```python
ytrain.value_counts()
```

```python
ytest.value_counts()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Create a neural network

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.callbacks import EarlyStopping
```

```python
model=Sequential()

model.add(Dense(64,activation='relu',input_dim=xtrain.shape[1]))
model.add(Dropout(0.3))

model.add(Dense(32,activation='relu'))
model.add(Dropout(0.25))

model.add(Dense(3,activation='softmax'))
```

```python
model.summary()
```

### Compile the model

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
early_stop=EarlyStopping(monitor='val_loss',patience=10)
```

```python
hist=model.fit(xtrain,ytrain_cat,validation_split=0.2,epochs=200,callbacks=[early_stop])
```

### Plot the learning curve

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(16,8))
plt.plot(hist.history['loss'],label='Training Losss')
plt.plot(hist.history['val_loss'],label='Validate Loss')
plt.xlabel('Epochs')
plt.ylabel('Cateogrical Cross Entropy')
plt.title('Learning curve')
plt.show()

```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(16,8))
plt.plot(hist.history['accuracy'],label='Training Accuarcy')
plt.plot(hist.history['val_accuracy'],label='Validate Accuarcy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy curve')
plt.show()

```

### Evaluate model accuracy

```python
model.evaluate(xtrain,ytrain_cat)
```

```python
model.evaluate(xtest,ytest_cat)
```

### Predicitin the results

```python
yprob_tr=model.predict(xtrain)
yprob_ts=model.predict(xtest)
```

```python
yprob_tr[0:5]
```

```python
yprob_ts[0:5]
```

### Convert probablity to prediction

```python
import numpy as np 
ypred_tr=np.argmax(yprob_tr,axis=-1)
ypred_ts=np.argmax(yprob_ts,axis=-1)
```

```python
ypred_tr[0:5]
```

```python
ypred_ts[0:5]
```

#### PLOT THE CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cf=confusion_matrix(ytest,ypred_ts)
cfd=ConfusionMatrixDisplay(cf)
cfd.plot()
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred_ts))
```
