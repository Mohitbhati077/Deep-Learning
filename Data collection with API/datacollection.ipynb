## Data Collection with API


```python
import requests

url = "https://amazon-product-price-data.p.rapidapi.com/product"

querystring = {"asins":"B005YQZ1KE,B074R8RQQ2,B0BLW25MB7,B088N8TF64","locale":"US"}

headers = {
	"X-RapidAPI-Key": "e1863bc0d5mshf04242d4c393a90p1ee066jsn714267894e89",
	"X-RapidAPI-Host": "amazon-product-price-data.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
data=response.json()
data
```

```python
data[0]
```

```python
data[0]['asin']
```

```python
data[0]['current_price']
```

```python
for i in data:
    print(i)
    
```

```python
asin=[]
product_name=[]
price=[]
for i in data:
    asin.append(i['asin'])
    product_name.append(i['product_name'])
    price.append(i['current_price'])
    
```

```python
asin
```

```python
price
```

```python
product_name
```

### Convert above in data frame

```python
dct={'asin':asin,
    'product_name':product_name,
    'price':price}
```

```python
import pandas as pd
df=pd.DataFrame(dct)
df
```
